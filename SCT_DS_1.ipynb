```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Other', 'Female']}
df = pd.DataFrame(data)

# Count the frequency of each gender
gender_count = df['Gender'].value_counts()

# Plot bar chart
gender_count.plot(kind='bar', color='skyblue')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [23, 25, 31, 22, 28, 33, 25, 29, 30, 24, 22, 28]}
df = pd.DataFrame(data)

# Plot histogram
plt.hist(df['Age'], bins=5, color='lightgreen', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
